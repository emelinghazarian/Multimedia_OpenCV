```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/faces.jpg')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```

```python
#using some pre-trained classifiers that have been provided by OpenCV Stored as .XML files
face_cascade= cv2.CascadeClassifier('haarcascades_haarcascade_frontalface_default.xml')

faces=face_cascade.detectMultiScale(gray_image,1.2, 6)

for (x, y, w, h) in faces:
    cv2.rectangle(image,(x,y), (x+w, y+h), (255,0,0), 3)
    
imshow("detected faces", image)
```
