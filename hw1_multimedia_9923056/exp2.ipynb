```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/img2.png')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)
```

```python
B, G, R = cv2.split(image)
imshow("Blue Channel Only", B)
imshow("Green Channel Only", G)
imshow("Red Channel Only", R)
```

```python
img_scaled1 = cv2.resize(B, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
imshow("2x Scaling - Inter Cubic", img_scaled1)
```

```python
img_scaled2 = cv2.resize(G, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
imshow("2x Scaling - Inter Cubic", img_scaled2)
```

```python
img_scaled3 = cv2.resize(R, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
imshow("2x Scaling - Inter Cubic", img_scaled3)
```

```python
B, G, R = cv2.split(image)
img_scaled1 = cv2.resize(B, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
img_scaled2 = cv2.resize(G, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
img_scaled3 = cv2.resize(R, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
merged = cv2.merge([img_scaled1,img_scaled2,img_scaled3]) 
imshow("Merged", merged) 
```

```python
hist1 = cv2.calcHist([merged],[0],None,[256],[0,256])
hist2 = cv2.calcHist([merged],[1],None,[256],[0,256])
hist3 = cv2.calcHist([merged],[2],None,[256],[0,256])
plt.plot(hist1), plt.plot(hist2),plt.plot(hist3)
plt.xlim([0,256])

plt.show()
```
