```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/img4.jpg')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype('uint8'))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

bilateral_filter = cv2.bilateralFilter(gray_image, 5, 45, 45)
imshow("bilateral filter", bilateral_filter)

gaussian_filter = cv2.GaussianBlur(gray_image, (5,5), 0)
imshow("gaussian filter", gaussian_filter)

median_filter = cv2.medianBlur(gray_image, 5)
imshow("median filter", median_filter)
```

```python
sobelx = cv2.Sobel(src=gaussian_filter, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=1)
sobely = cv2.Sobel(src=gaussian_filter, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=1)
sobelxy = cv2.Sobel(src=gaussian_filter, ddepth=cv2.CV_32F, dx=1, dy=1, ksize=1)

imshow('Sobel X', sobelx)
imshow('Sobel Y', sobely)
imshow('Sobel XY_gaussian', sobelxy)

# Canny Edge Detection
edges = cv2.Canny(image=gray_image, threshold1=100, threshold2=200)
imshow('Canny_gaussian', edges)
```

```python
sobelx = cv2.Sobel(src=bilateral_filter, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=1)
sobely = cv2.Sobel(src=bilateral_filter, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=1)
sobelxy = cv2.Sobel(src=bilateral_filter, ddepth=cv2.CV_32F, dx=1, dy=1, ksize=1)

imshow('Sobel X', sobelx)
imshow('Sobel Y', sobely)
imshow('Sobel XY_bilateral', sobelxy)

# Canny Edge Detection
edges = cv2.Canny(image=gray_image, threshold1=100, threshold2=200)
imshow('Canny_bilateral', edges)
```

```python
sobelx = cv2.Sobel(src=median_filter, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=1)
sobely = cv2.Sobel(src=median_filter, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=1)
sobelxy = cv2.Sobel(src=median_filter, ddepth=cv2.CV_32F, dx=1, dy=1, ksize=1)

imshow('Sobel X', sobelx)
imshow('Sobel Y', sobely)
imshow('Sobel XY_median', sobelxy)

# Canny Edge Detection
edges = cv2.Canny(image=gray_image, threshold1=100, threshold2=200)
imshow('Canny_median', edges)
```
