```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/img3.jpg')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)
```

```python
kernel_circular=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(9,9))

#img_temp1=cv2.erode(image, kernel_circular,iterations=1)
circular_img=cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel_circular)
imshow("circles", circular_img)
```

```python
img_temp1=cv2.erode(circular_img, kernel_circular,iterations=1)
imshow("test", img_temp1)
```

```python
kernel_circular_test=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4))
img_temp2=cv2.morphologyEx(img_temp1,cv2.MORPH_OPEN,kernel_circular_test)
imshow("test2", img_temp2)
```

```python
img_final=cv2.dilate(img_temp2,kernel_circular,iterations=1)
imshow("final", img_final)
```

```python
new_image=(image)-(img_final)
imshow('without circles', new_image)
```

```python
kernel_new=cv2.getStructuringElement(cv2.MORPH_RECT,(4,4))
new_image_temp=cv2.erode(new_image, kernel_new,iterations=1)
imshow("temp_test", new_image_temp)
```

```python
img_final_whithout_circles=cv2.dilate(new_image_temp,kernel_new,iterations=1)
imshow("img_final_whithout_circles",img_final_whithout_circles )
```
