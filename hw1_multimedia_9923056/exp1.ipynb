```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/img1.tif')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imgnew3= 255 - (gray_image)
imshow("first trans", imgnew3)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image2=gray_image
for i in range(150):
    for j in range(150):
        if gray_image[i][j]>255*0.2 and gray_image[i][j]<255*0.55:
            image2[i][j]=0
imshow("second trans",image2)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image3=gray_image
for i in range(150):
    for j in range(150):
        if gray_image[i][j]<255*0.4:
            image3[i][j]=0
        elif gray_image[i][j]>255*0.55:
            image3[i][j]=255
imshow("third trans",image3)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image4=gray_image
for i in range(150):
    for j in range(150):
        if gray_image[i][j]<255*0.2:
            image4[i][j]=1.5*gray_image[i][j]
        elif gray_image[i][j]>255*0.2 and gray_image[i][j]<255*0.55:
            image4[i][j]=0.3*255
        elif gray_image[i][j]>255*0.55:
            image4[i][j]=(0.7/0.45)*gray_image[i][j]+((1-(0.7/0.45))*225)
imshow("fourth trans",image4)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image6=gray_image
for i in range(150):
    for j in range(150):
        image6[i][j]=((((1/255)*gray_image[i][j])**2)*255)
imshow("sixth trans",image6)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image5=gray_image
for i in range(150):
    for j in range(150):
        image5[i][j]=((((1/255)*gray_image[i][j])**(1/2))*255)
imshow("fifth trans",image5)
```
