```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('C:/Users/Emelin/Desktop/images/img3.jpg')

def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Displaying Our First Image", image)
```

```python
#kernel_length=np.array(image).shape[1]//8

vertical_kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(1,20))

hori_kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(20,1))

kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(1,1))

img_temp1=cv2.erode(image, vertical_kernel,iterations=1)
vertical_lines_img=cv2.morphologyEx(img_temp1,cv2.MORPH_OPEN,vertical_kernel)
imshow("vertical lines", vertical_lines_img)

img_temp2=cv2.erode(image, hori_kernel,iterations=1)
hori_lines_img=cv2.morphologyEx(img_temp2,cv2.MORPH_OPEN,hori_kernel)
imshow("hori lines", hori_lines_img)
```
